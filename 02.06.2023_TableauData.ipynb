```python
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandasql import sqldf
```

```python
conn = sqlite3.connect(r'C:\Users\Britt\AppData\Roaming\DBeaverData\workspace6\.metadata\sample-database-sqlite-1\Chinook.db')
```

```python
artist_data = pd.read_sql_query('''SELECT a.Name as Artist_Name, a2.Title as Album_Title, g.Name as Genre, t.Composer as Composer, t.Name as Track_Name, t.UnitPrice as Track_Unit_Price, il.InvoiceId as Invoice_ID, il.UnitPrice as Invoice_Tracks_Price, il.TrackId as Invoice_Track_ID, e.EmployeeId as Sales_Rep_ID, e.FirstName as Sales_Rep_First_Name, e.LastName as Sales_Rep_Last_Name, e.Title as Sales_Rep_Title, e.City as Sales_Rep_City, e.State as Sales_Rep_State, e.Country as Sales_Rep_Country, c.CustomerId as Customer_ID, c.FirstName as Customer_First_Name, c.LastName as Customer_Last_Name, c.City as Customer_City, c.State as Customer_State, c.Country as Customer_Country
FROM Artist a 
LEFT JOIN Album a2 on a2.ArtistId = a.ArtistId
LEFT JOIN Track t on t.AlbumId = a2.AlbumId 
LEFT JOIN Genre g on g.GenreId = t.GenreId 
LEFT JOIN InvoiceLine il on il.TrackId = t.TrackId 
LEFT JOIN Invoice i on i.InvoiceId = il.InvoiceId 
LEFT JOIN Customer c on c.CustomerId = i.CustomerId 
LEFT JOIN Employee e on e.EmployeeId = c.SupportRepId''', conn)
conn.close()
```
